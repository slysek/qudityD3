```python
from qiskit import qpy
from qiskit.circuit import QuantumCircuit
from qiskit.quantum_info import Statevector, Operator
import numpy as np
import matplotlib.pyplot as plt
from qiskit.quantum_info import partial_trace
```

```python
with open('Xgate4.qpy', 'rb') as fd:
    Xgate4 = qpy.load(fd)[0]

with open('Zgate4.qpy', 'rb') as fd:
    Zgate4 = qpy.load(fd)[0]

with open('Fgate4.qpy', 'rb') as fd:
    Fgate4 = qpy.load(fd)[0]

with open('CZgate4.qpy', 'rb') as fd:
    CZgate4 = qpy.load(fd)[0]
```

```python
ame44 = QuantumCircuit(8)

for i in range(4):
    ame44.append(Fgate4, [2*i, 2*i + 1])

ame44.append(CZgate4, [0, 1, 2, 3])
ame44.append(CZgate4, [2, 3, 4, 5])
ame44.append(CZgate4, [4, 5, 6, 7])
ame44.append(CZgate4, [4, 5, 6, 7])
ame44.append(CZgate4, [6, 7, 0, 1])

ame44.draw('mpl')
```

```python
Statevector(ame44).draw('latex')
```

```python
partial_trace(Statevector(ame44), [0, 1, 2, 3, 4, 5]).draw('latex')
```

```python
partial_trace(Statevector(ame44), [0, 1, 2, 3]).draw('latex')
```

```python
np.diag(partial_trace(Statevector(ame44), [0, 1, 2, 3]).data).round(3) 
```
